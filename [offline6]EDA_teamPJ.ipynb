```python
# sale_price 결측치, transaction_amount 절대값으로 fillna
df['sale_price'] = df['sale_price'].fillna(df['transaction_amount'].abs())
```

```python
# sale_price 이상치 처리(sale_price 75%(IQR3) * 3 이상 drop)
# sale_price 이상치 중에 작은 값제거 X (min fence)
# sale_price 중에서 이상치가 하단부의 이상치는 사용할 것이고, 상단부는 사용하지 않기 때문에,
def remove_out(dataframe, remove_col):
    dff = dataframe
    for k in remove_col:
        level_1q = dff[k].quantile(0.25)
        level_3q = dff[k].quantile(0.75)
        IQR = level_3q - level_1q
        rev_range = 3  # 제거 범위 조절 변수
        dff = dff[dff[k] <= level_3q + (rev_range * IQR)] # 해당의 코드로 상단부 이상치만 drop하고 나머지만 활용하기 위하여 처리
        dff = dff.reset_index(drop=True)
    return dff
df = remove_out(df, ['sale_price'])
```

### type 과 state 에 대한 정리_오세연
- ORDER 의 COMPLETED 여야 함에도, 이전의 데이터의 처리 과정에 따라 21년 7월 이전의 ORDER의 내역 중 ORDER COMPLETED 가 아니라, ORDER CANCELLED 처리가 된 데이터들이 있어, 해당의 기준을 ORDER_COMPLETED 와 PAYMENT_COMPLETED 가 같은 강의로 한 사람의 사용자가 같은 시간대에 이용한 것을 기반으로 하여, 조건식에 따라 ORDER의 state 상태를 정리
- 사유 : ORDER 라는 상태 자체가 고객이 구매에 대한 고민을 하거나 구매에 대한 결정을 했다는 상태를 보여주기 때문.

```python
# copy the df just in case & not to corrupt data
test = df.copy()
test # 99068 rows × 17 columns
```

```python
# conditions for intersecting datasets -> 조건(ORDER_CANCELLED 와 PAYMENT_COMPLETED)
odrcan = test[(test['type'] == 'ORDER') & (test['state'] == 'CANCELLED')] # 34121 rows × 17 columns
paycom = test[(test['type'] == 'PAYMENT') & (test['state'] == 'COMPLETED')] # 51706 rows × 17 columns
```

```python
# order_cancelled 이면서, payment_completed인 사람들 바꿔야 할 대상.
inter = odrcan[odrcan['customer_id'].isin(paycom['customer_id']) == True]
inter # 21996 rows × 17 columns
```

```python
# 위의 조건에 맞는 대상들의 id 를 기반으로, ORDER의 state 조정
# change wrong order_cancelled to order_completed
for t, testId in test.id.iteritems():
    for i, interId in inter.id.iteritems():
        if testId == interId : 
            if test['type'][t] == 'ORDER' : 
                # test['type'][t] = 'ORDER'
                test['state'][t] = 'COMPLETED'
print('done')
```
